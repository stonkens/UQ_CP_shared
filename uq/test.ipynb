```python
import numpy as np
```

```python
center = np.array([0.0, 0.0])
radii = 1.0
```

```python
sdf = lambda x: np.linalg.norm(x - center, axis=-1) - radii
squared_sdf = lambda x: np.linalg.norm(x - center, axis=-1)**2 - radii**2
```

```python
# Generate (-5, 5) x (-5, 5) grid for x, y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

all_eval_pts = np.concatenate([X[..., np.newaxis], Y[..., np.newaxis]], axis=-1)


Z1 = sdf(all_eval_pts)
Z2 = squared_sdf(all_eval_pts)
```

```python
Z1.shape
```

```python
import matplotlib.pyplot as plt
contours = plt.contourf(X, Y, Z1)
plt.colorbar(contours)
```

```python
contours = plt.contourf(X, Y, Z2-Z1)
# Draw legend 
plt.colorbar(contours)
```
